### Gradient Descent

````python
# Define the function f(x) and its derivative df(x)
def f(x):
    return x**2  # Example function, you can replace it with your function

def df(x):
    return 2*x  # Derivative of x^2

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]  # Initialize with the derivative at the starting point

for n in range(500):
    previous_x = new_x
    gradient = df(new_x)  # Calculate the gradient (derivative at new_x)
    new_x = previous_x - step_multiplier * gradient  # Update new_x using gradient descent formula

    step_size = abs(new_x - previous_x)

    x_list.append(new_x)
    slope_list.append(df(new_x))  # Append the derivative (df(x)) at the new_x

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

### Example 3 - Divergence and Overflow

#### $$h(x) = x^5 - 2x^4 + 2$$

````python
## TASK-4 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt

# Define the function h(x) and its derivative dh(x)
def h(x):
    return x**5 - 2*x**4 + 2  # Function h(x) = x^5 - 2x^4 + 2

def dh(x):
    return 5*x**4 - 8*x**3  # Derivative dh(x) = 5x^4 - 8x^3

# Gradient Descent
new_x = 3  # Starting point
previous_x = 0
step_multiplier = 0.01  # Step size for gradient descent
precision = 0.00001  # Precision for stopping the gradient descent

x_list = [new_x]
slope_list = [dh(new_x)]  # Derivative at the starting point

# Running gradient descent to minimize h(x)
for n in range(500):
    previous_x = new_x
    gradient = dh(new_x)  # Calculate the gradient (derivative at new_x)
    new_x = previous_x - step_multiplier * gradient  # Update new_x using gradient descent formula

    step_size = abs(new_x - previous_x)

    x_list.append(new_x)
    slope_list.append(dh(new_x))  # Append the derivative (dh(x)) at the new_x

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or dh(x) value at this point is:', dh(new_x))
print('h(x) value or cost at this point is:', h(new_x))

# Plotting the function h(x) and its derivative dh(x)
x_vals = np.linspace(-2, 3, 1000)  # Values of x for plotting

plt.figure(figsize=(12, 6))

# Plot the function h(x) on the left
plt.subplot(1, 2, 1)  # (rows, columns, plot number)
plt.plot(x_vals, h(x_vals), label="h(x) = x^5 - 2x^4 + 2", color='blue')
plt.title("Function: h(x) = x^5 - 2x^4 + 2")
plt.xlabel('x')
plt.ylabel('h(x)')
plt.grid(True)
plt.legend()

# Plot the derivative dh(x) on the right
plt.subplot(1, 2, 2)  # (rows, columns, plot number)
plt.plot(x_vals, dh(x_vals), label="dh(x)/dx = 5x^4 - 8x^3", color='red')
plt.title("Derivative: dh(x)/dx = 5x^4 - 8x^3")
plt.xlabel('x')
plt.ylabel('dh(x)/dx')
plt.grid(True)
plt.legend()

# Show the plots side by side
plt.tight_layout()
plt.show()
````

````python
## TASK-7 : Plot reduction in cost for each iteration
import numpy as np
import matplotlib.pyplot as plt

# Define the function g(x) and its derivative dg(x)
def g(x):
    return x**2  # Example function (for g(x))

def dg(x):
    return 2 * x  # Derivative of g(x) = x^2

# Gradient Descent function with cost tracking
def gradient_descent(derivative_func, cost_func, initial_guess, multiplier, precision, max_iter):
    new_x = initial_guess
    previous_x = 0
    x_list = [new_x]
    cost_list = [cost_func(new_x)]  # Store cost (function value) at each iteration

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(new_x)  # Calculate the gradient (derivative at new_x)
        new_x = previous_x - multiplier * gradient  # Update new_x using gradient descent formula

        # Track the cost (function value) at each iteration
        cost_list.append(cost_func(new_x))

        # Calculate the step size (change in x) to check for convergence
        step_size = abs(new_x - previous_x)

        if step_size < precision:
            print(f"Loop ran this many times with multiplier {multiplier}:", n)
            break

    return x_list, cost_list

# Run gradient descent
n = 100
low_gamma_x, low_gamma_cost = gradient_descent(derivative_func=dg, cost_func=g, initial_guess=3,
                                                multiplier=0.0005, precision=0.0001, max_iter=n)

mid_gamma_x, mid_gamma_cost = gradient_descent(derivative_func=dg, cost_func=g, initial_guess=3,
                                                multiplier=0.001, precision=0.0001, max_iter=n)

high_gamma_x, high_gamma_cost = gradient_descent(derivative_func=dg, cost_func=g, initial_guess=3,
                                                multiplier=0.002, precision=0.0001, max_iter=n)

# Plotting the cost reduction for low, mid, high gamma values
plt.figure(figsize=(14, 8))

# Plot cost reduction for low gamma
plt.subplot(2, 2, 1)
plt.plot(low_gamma_cost, label="Low gamma (0.0005)", color='blue')
plt.title("Cost Reduction with Low Learning Rate")
plt.xlabel('Iterations')
plt.ylabel('Cost (g(x))')
plt.grid(True)
plt.legend()

# Plot cost reduction for mid gamma
plt.subplot(2, 2, 2)
plt.plot(mid_gamma_cost, label="Mid gamma (0.001)", color='green')
plt.title("Cost Reduction with Mid Learning Rate")
plt.xlabel('Iterations')
plt.ylabel('Cost (g(x))')
plt.grid(True)
plt.legend()

# Plot cost reduction for high gamma
plt.subplot(2, 2, 3)
plt.plot(high_gamma_cost, label="High gamma (0.002)", color='red')
plt.title("Cost Reduction with High Learning Rate")
plt.xlabel('Iterations')
plt.ylabel('Cost (g(x))')
plt.grid(True)
plt.legend()

# Show all plots
plt.tight_layout()
plt.show()
````

#### Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
import numpy as np

# Example data (replace this with your actual data)
x_5 = np.array([1, 2, 3, 4, 5])  # Features
y_5 = np.array([1, 2, 1.3, 3.75, 2.25])  # True values

# Mean Squared Error (MSE) function
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Gradient of MSE function with respect to thetas (theta_0 and theta_1)
def grad(x, y, thetas):
    theta_0, theta_1 = thetas
    m = len(x)
    y_pred = theta_0 + theta_1 * x
    error = y_pred - y
    d_theta_0 = (2/m) * np.sum(error)  # Gradient w.r.t. theta_0
    d_theta_1 = (2/m) * np.sum(error * x)  # Gradient w.r.t. theta_1
    return np.array([d_theta_0, d_theta_1])

# Gradient Descent to minimize MSE
multiplier = 0.01
thetas = np.array([2.9, 2.9])  # Initial values for theta_0 and theta_1

# Collect data points for plotting
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

# Gradient descent iterations
for i in range(1000):
    gradients = grad(x_5, y_5, thetas)  # Get the gradients
    thetas = thetas - multiplier * gradients  # Update thetas

    # Store new values for plotting
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(mse_vals, mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
````
